# Aula 1 - Equações não lineares

Aqui aprenderemos algumas técnicas que nos ajudarão a encontrar *aproximações* para as raízes isoladas de um polinômio ou função transcendente `f(x)`.

> **Teorema 1:** Se uma função contínua `f(x)` assume valores de sinais opostos nos pontos extremos de um intervalo dado por `[a, b]`, ou seja, se `f(a) x f(b) < 0`, então existe pelo menos um ponto `x' ∈ [a, b]`, tal que `f(x') = 0`.



## Método da Bissecção

O método da bissecção consiste em, através do uso do **Teorema 1**, diminuir o comprimento dos intervalos dentros dos quais há raízes da função. O método é, basicamente:

1. Tome f(x) como uma função contínua em `[a, b]`;
2. Seccione o intervalo ao meio, tomando um x<sub>1</sub> = `(a + b) / 2`;
3. Agora, calculemos f(x<sub>1</sub>) e analisemos seu resultado:
    1. Se f(x<sub>1</sub>) = 0, então este é o melhor caso, e encontramos *exatamente* uma raiz de f;
    2. Se f(a) x f(x<sub>1</sub>) < 0, então sabemos que há uma raiz entre a e x<sub>1</sub> e devemos executar uma nova bissecção em [a, x<sub>1</sub>];
    3. Se f(a) x f(x<sub>1</sub>) > 0, então podemos deduzir que f(x<sub>1</sub>) x f(b) < 0, compelindo uma nova bissecção em [x<sub>1</sub>, b];

Através desse método, nós vamos nos aproximando da raiz, mas como saber quando parar a execução? Como identificar uma aproximação suficiente?

## Condição de parada

Para sabermos identificar uma boa aproximação, e portanto a condição de parada dos algoritmos, precisamos entender o conceito de **erro**. 

### Erro absoluto

O erro absoluto de uma dada iteração pode ser definido como |f(x<sub>k+1</sub>) - f(x<sub>k</sub>)| < ε, mas deve ser usado com cuidado, pois o erro absoluto pode significar precisões muito diferentes a depender do comportamento da função (1.000.001 - 1.000.000 = 1 tal como 2 - 1 = 1, mas no segundo caso o erro é de 100%).

## Erro relativo

O erro relativo considera uma relação mais contextualizada, definida por: $${|f(x_{k+1}) - f(x_k)| \over {f(x_{k+1})}} < ε$$

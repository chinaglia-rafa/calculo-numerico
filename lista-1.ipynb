```python
import numpy as np
import matplotlib.pyplot as plotter
```

```python
def bissec(f, a, b, tol = 1e-6, maxIt = 50):
    # Veririca se há raiz entre a e b
    if f(a) * f(b) > 0:
        return None, 0
    
    # Verifica se as extremidades a ou b são aproximações suficientes
    if np.abs(f(a)) < tol:
        return a, 0
    if np.abs(f(b)) < tol:
        return b, 0
    
    for it in range(maxIt):
        x = (a + b) / 2

        if np.abs(f(x)) < tol:
            return x, it+1
        
        if f(a) * f(x) < 0:
            b = x
        else:
            a = x
    print("Número máximo de iterações atingidos!")
    return x, maxIt
```

a) $x - 2^{-x} = 0$, com $0\leq x\leq1$

```python
f_1_a  = lambda x: x - 2**(-x)

x, it = bissec(f_1_a, 0, 1, 1e-6, 50)

print(x, it)

pts = np.linspace(0, 1, 100)
plotter.plot(pts, f_1_a(pts))
plotter.hlines(0, 0, 1, 'k')
plotter.scatter([x], [0], color="red")
```

b) $exp(x) - x^2 + 3x -2 = 0$ no intervalo $0\leq x \leq 1$

```python
f_1_b = lambda x: np.exp(x) - x*x + 3*x - 2

x, it = bissec(f_1_b, 0, 1)

print(x, it)

pts = np.linspace(0, 1, 100)
plotter.plot(pts, f_1_b(pts))
plotter.hlines(0, 0, 1, 'k')
plotter.scatter([x], [0], color="red")
```

c) $2x * cos(2x) - (x + 1)^2 = 0$ nos intervalos $-3 \leq x \leq -2$ e $-1 \leq x \leq 0$

```python
f_1_c = lambda x: 2*x * np.cos(2 * x) - (x + 1)**2

print("Para [-3, -2]")
x1, it1 = bissec(f_1_c, -3, -2)
print(x1, it1)

print("Para [-1, 0]")
x2, it2 = bissec(f_1_c, -1, 0)
print(x2, it2)

pts = np.linspace(-3, 0, 100)
plotter.plot(pts, f_1_c(pts))
plotter.hlines(0, -3, 0, 'k')
plotter.scatter([x1], [0], color="red")
plotter.scatter([x2], [0], color="cyan")
```

d) $x * cos(x) - 2 * x^2 + 3x - 1 = 0$ nos intervalos $-0.3 \leq x \leq 0.2$ e $1.2 \leq x \leq 1.3$

```python
f_1_d = lambda x: x * np.cos(x) - 2 * x*x + 3 * x - 1

print("Para [-0.3, 0.2]")
x1, it1 = bissec(f_1_d, -0.3, 0.2)
print(x1, it1)
print("Para [1.2, 1.3]")
x2, it2 = bissec(f_1_d, 1.2, 1.3)
print(x2, it2)

pts = np.linspace(-3, 1.3, 100)
plotter.plot(pts, f_1_d(pts))
plotter.hlines(0, -3, 1.3, 'k')
plotter.scatter([x1], [0], color="red")
plotter.scatter([x2], [0], color="cyan")
```

```python
def newton(f,df, x, tol = 1e-6, maxIt=50):
    for it in range(maxIt):
        if np.abs(f(x)) < tol:
            return x, it
        if df(x) == 0:
            print("Erro: df(x) = 0")
            return x, it
            # Alternativamente:
            # raise ValueError(f"A derivada em x = {x} é 0")
        x = x - f(x)/df(x)
    print('Numero max de iterações foi atingido')
    return x, maxIt
```

a) $exp(x) + 2^{-x} + 2 * cos(x) - 6 = 0$ no intervalo $1 \leq x \leq 2$

```python
x = 1.5
f = lambda x: np.exp(x) + 2 ** (-x) + 2 * np.cos(x) - 6
df = lambda x: np.exp(x) - 2 ** (-x) * np.log(2) - 2 * np.sin(x)
val, it = newton(f, df, x)

print(val, it)
pts = np.linspace(-4, 2, 100)
plotter.plot(pts, f(pts))
plotter.plot(pts, df(pts))
plotter.hlines(0, -4, 2, 'k')
plotter.scatter([val], [0], color="red")


```
